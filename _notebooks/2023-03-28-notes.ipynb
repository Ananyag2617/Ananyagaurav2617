# Scholarships
- Teachers can assist students with their essays or application materials.
- The quantum bridge program, summer scholars, and other initiatives can aid in determining a professional path.
- Scholarships and grants are both privately sponsored and based on demographic and financial factors.
- In addition to applying for scholarships and grants during college, loans might be subsidized or unsubsidized.
- Following graduation from college, one may pursue a career in education, government service, or the biochemical, environmental, chemical, or electronic fields.
- Subsidized and unsubsidized loans are both types of loans that are offered to students by the US Department of Education as a part of the Federal Direct Loan Program.
- Subsidized loans are available to undergraduate students who demonstrate financial need. The government pays the interest on the loan while the student is in school, as well as during the six-month grace period after graduation, and during any periods of deferment. This means that the borrower will only be responsible for paying back the principal amount of the loan, without accruing interest during these periods.
- Unsubsidized loans are available to undergraduate and graduate students regardless of their financial need. With unsubsidized loans, the borrower is responsible for paying all of the interest that accrues on the loan, including while they are in school, during the grace period, and any periods of deferment. The borrower can choose to make interest payments while they are in school, or allow the interest to accrue and be added to the principal balance of the loan.
